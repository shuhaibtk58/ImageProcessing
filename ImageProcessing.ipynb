```python
blue_channel, green_channel, red_channel = cv2.split(image)
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
start_point = (50, 50)
end_point = (200, 200)
color = (255, 0, 0)  # Blue color

image_with_line = cv2.line(image, start_point, end_point, color, thickness=2)

arrow_start = (300, 100)
arrow_end = (450, 150)
image_with_arrow = cv2.arrowedLine(image_with_line, arrow_start, arrow_end, color, thickness=2, tipLength=0.4)

cv2.imshow('Image with Line and Arrow', image_with_arrow)
cv2.waitKey(0)  
cv2.destroyAllWindows()  
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'image.png'  
image = cv2.imread(image_path)

cv2.imshow('Image', image)
cv2.waitKey(0)  
cv2.destroyAllWindows() 
```

```python
image.shape
```

```python
image = cv2.resize(image, (500, 500))
cv2.imshow('Image', image)
cv2.waitKey(0)  
cv2.destroyAllWindows() 
```

```python
image.shape
```

```python
angle = 45
height = 500
width = 500
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))         #Rotation

cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rectangle = cv2.rectangle(image, (0, 0), (100, 150), (255, 0, 0), 2)        #Rectangle

cv2.imshow('ot', image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
center_coordinates = (50, 50)
radius = 50
color = (0, 255, 0)  # Green color in BGR
thickness = 2

cv2.circle(image, center_coordinates, radius, color, thickness)          #Circle

cv2.imshow('Circle', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
center_coordinates = (100, 50)
axes_lengths = (100, 50)
angle = 0
start_angle = 0
end_angle = 360
color = (0, 255, 255)  # Yellow color in BGR
thickness = 2

cv2.ellipse(image, center_coordinates, axes_lengths, angle, start_angle, end_angle, color, thickness)          #Ellipse

cv2.imshow('Ellipse', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
text = 'Hello, World!'
org = (50, 50)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (0, 0, 0)  # White color in BGR
thickness = 2

cv2.putText(image, text, org, font, font_scale, color, thickness)             #Text

cv2.imshow('Image with Text', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
added_image = cv2.add(image1, image2)

cv2.imshow('Added Image', added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
subtracted_image = cv2.subtract(image1, image2)

cv2.imshow('Subtracted Image', subtracted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(0)
while True:
    success, frame = video.read()
    cv2.imshow('video', frame)
    
    if cv2.waitKey(1) & 0xFF == 27:  # Press 'Esc' ke to break the loop
        break

video.release()
cv2.destroyAllWindows()
```
